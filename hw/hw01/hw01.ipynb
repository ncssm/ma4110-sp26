## 2. Analyzing Text in *Pride and Prejudice*

We can use Python to analyze the classic novel [*Pride and Prejudice*](https://www.gutenberg.org/ebooks/1342) by Jane Austen. The following code generates a scatter plot in which each dot corresponds to a chapter of the book. **Don't worry about how this code works for now! Just run it to see the scatter plot it produces**

````python
# Just run this cell
from datascience import *
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

# Read the text of Pride and Prejudice
with open('pride.txt', 'r') as file:
    pride_text = file.read()

# Split into chapters (the book uses "CHAPTER" as delimiter)
chapters = pride_text.split('CHAPTER ')[1:]  # Skip text before first chapter

# Count sentences (periods) and words in each chapter
sentences = [chapter.count('.') for chapter in chapters]
words = [len(chapter.split()) for chapter in chapters]

Table().with_columns(
    'Sentences',  sentences,
    'Words', words
    ).iscatter('Sentences')
````
